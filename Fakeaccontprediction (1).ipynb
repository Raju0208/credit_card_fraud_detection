This dataset is highly unbalanced

0 ---> Normal Transaction          1  ---> fraudulent transaction




```python
# seperating the data for the analysis
# here we classify the dataset in two variabl seperatily in fraud and ligit
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

Under_Sampling

Build a sample dataset containing similar distribution of normal transactions ans Fraudulent transactions

Number of fraud transaction = 88

```python
# here we store all (88) legit data in legit_sample variable
legit_sample = legit.sample(n=479*8)
```

```python
legit_sample.shape
```

```python
fraud
```

```python
fraud.shape
```

Concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample , fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset.shape
```

Spliting the data into Features & Targets

```python
# now we have to predict the class ( class = 1-> fraud and =0-> legit) so store class in y and rest of all put in the x varible (by dropping class from the data set)
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
print(Y)
```

Split the data into train data and the test data

```python
# divide the dataset for testing and training , here we use 20% data for test and 80% data for train
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , stratify=Y , random_state=42)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model training

Logistics Regression

```python
# Now from here we start prediction with Logistics regression
model = LogisticRegression()
```

```python
# training the Logistic regression Model with training data , so that we can test the model
model.fit(X_train , Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy )
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy )
```

## f1 score

```python
from sklearn.metrics import f1_score
f1 = f1_score(Y_test, X_test_prediction, average='binary')
f1
```

## Decision tree


```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train , Y_train)
```

```python
# accuracy on training data
X_train_prediction = clf.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
print('Accuracy on training data : ',training_data_accuracy )
```

```python
#accuracy on test data
X_test_prediction = clf.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data : ',test_data_accuracy )
```
